```python
import pandas as pd
import os
import sys
import time
import matplotlib.pyplot as plt
import scipy.signal
import numpy as np
import seaborn as sns
import datetime
```

```python
raw_lul=pd.read_csv('LUL_IMU11.csv', usecols=['time', 'ax', 'az'])
```

```python
raw_lul['time'] =  pd.to_datetime(raw_lul['time'])
raw_lul=raw_lul.rename(columns={'time':'timestamp'})
raw_lul['ist_time']=raw_lul['timestamp']+datetime.timedelta(hours=5, minutes=30)
```

```python
raw_lul
```

```python
%matplotlib
```

```python
raw_lul["acc"] = ( raw_lul["ax"]**2 + raw_lul["az"]**2 ) ** 0.5
```

```python
# plt.plot(raw_lul['time'],raw_lul['acc'])
```

```python
acc=raw_lul[["timestamp", "acc"]]
acc.set_index('timestamp')
test_lul=acc
```

```python
test_lul
```

```python
test_lul['rlg_acc']=pd.Series.to_frame(test_lul['acc'].rolling(100, center=True).std())
```

```python
test_lul
```

```python
timethresh=10
```

```python
import scipy.signal
import numpy as np
```

```python
plt.plot(test_lul['timestamp'],test_lul['rlg_acc'])
```

```python
board=scipy.signal.find_peaks(test_lul.rlg_acc, height=(0.02), distance=timethresh*100, width=1)
```

```python
board
```

```python
# plt.plot(raw_lul['time'],raw_lul['acc'])
```

```python
differ=np.diff(board[0])
differ=differ.tolist()
```

```python
differ
```

```python
for x in board[0]:
    test_lul.at[x, 'state'] = 1

lul_boards=test_lul[test_lul['state']==1]
print('Boards detected in LUL:', lul_boards.shape[0])
```

```python
fig, ax1 = plt.subplots()
# ax1.plot(test_ll['timestamp'], np.ones(test_ll.shape[0])*0.01, 'green')
ax1.plot(test_lul['timestamp'], test_lul['rlg_acc'], 'orange')

# ax1.plot(test_ld['timestamp'], np.ones(test_ld.shape[0]*0.01), 'green')
# ax1.xaxis.set_major_locator(mdates.DateFormatter('%H:%M'))

ax1.set_xlabel('Time')

ax2 = ax1.twinx()
ax2.stem(lul_boards['timestamp'], lul_boards['state'], 'b:')

plt.title('Board Detections')
```

```python
unloading_delays_rawdf=pd.DataFrame({"sample_number":board[0], "working_time":board[1]['widths']/100})
LUL_events=pd.DataFrame({"timestamp":test_lul.iloc[unloading_delays_rawdf.sample_number].timestamp, "event":1, "working_time":(unloading_delays_rawdf.working_time).tolist()})
```
